```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Draw a line Chart - Like a Race Track

```python
x = [1,2,3,4,5,6]
y=[10,20,30,30,35,38]
```

```python
plt.plot(x,y)
plt.title("Race Progress")
plt.xlabel("Time")
plt.ylabel("Spead")
plt.grid(True)
plt.show()
```

Bar Chart - Compare Chocolate Sales

```python
chocolates = ["Kitkat", "Perk", "Munch", "Diary Milk"]
sales = [82,90,70,94]
```

```python
plt.bar(chocolates,sales, color='purple')
plt.title("Chocolate Sales")
plt.ylabel("Bars Sold")
plt.show()
```

Pie Chart - What you ate at Buffet

```python
food = ['Pizza', "Burger", "Fries",'Ice Cream']
percent = [30,29,35,35]
```

```python
plt.pie(percent, labels=food, autopct='%1.1f%%')
plt.title("What I ate")
plt.show()
```

Seaborn - Fancy graphs

```python
#Load a sample Dataset (Built in)
```

```python
tips= sns.load_dataset("tips")
```

Plotting with seaborn

```python
sns.histplot(tips['total_bill'],kde= True)
plt.title("Total Bill Distribution")
plt.show()
```

Scatter Plot - Tip vs Bill

```python
sns.scatterplot(x='total_bill', y = 'tip', data=tips, hue='sex')
plt.title("Does Tip Increases with Bill? ")
plt.show()
```

Box Plot - Tip distribution by Gender

```python
sns.boxplot(x='sex', y = 'tip', data=tips)
plt.title("Tip Distribution by Gender")
plt.show()
```
