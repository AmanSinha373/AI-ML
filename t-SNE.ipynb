# t-SNE 

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler  # Standardize data first
from sklearn.datasets import load_digits # Digits dataset (64 features per sample)
```

# Load the dataset

```python
digits = load_digits()
X = digits.data
y= digits.target
```

# STEP 1 : Standard scaling ( important before tSNE)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# STEP 2 : Apply t-SNE

```python
tsne = TSNE(n_components=2, perplexity=30, n_iter = 300,random_state=42)
X_tsne= tsne.fit_transform(X_scaled)
```

# STEP 3 : PLOT

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_tsne[:,0], y = X_tsne[:,1], hue=y, palette="tab10", legend='full')
plt.title("t-SNE Visualization of Digits Dataset")
plt.show()
```
